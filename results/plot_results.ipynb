```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.ndimage.filters import gaussian_filter1d
sns.set_style('whitegrid')
```

```python
rewards = ['partial', 'sparse']
agents  = ['ground_truth', 'attn_mech', 'dyna', 'None']
labels  = ['CoDA (ground truth)', 'CoDA (learned)', 'Dyna (learned)', 'Base Agent (TD3)']
colors  = ['#e41a1c','#377eb8','#4daf4a','#984ea3','#ff7f00','#ffff33','#a65628','#f781bf']
COLOR   = dict(zip(agents, colors))
LABEL   = dict(zip(agents, labels))
#colors  = ['#66c2a5','#fc8d62','#8da0cb','#e78ac3','#a6d854','#ffd92f','#e5c494','#b3b3b3']
#colors  = ['#1b9e77','#d95f02','#7570b3','#e7298a','#66a61e','#e6ab02','#a6761d','#666666']
Ns      = [1,2,3,4]
seeds   = [0, 1, 2]
M       = 350
```

```python
def plot_result(reward, N, coda, ax = plt):
  
  # Load the result plots
  files = [f"TD3_place_{reward}_{N}_{coda}___{s}.npy" for s in seeds]
  plots = [np.load(file) for file in files]
  M     = np.min([len(p) for p in plots])
  plots = [p[:M] for p in plots]
  
  # plots the result plots
  c = COLOR[coda]
  [ax.plot(p, c=c, alpha=0.2) for p in plots]
  ax.plot(gaussian_filter1d(np.mean(plots, axis=0).copy(),sigma=10), c=c, linewidth=2)
  legend = ax.plot([], color=c, label=LABEL[coda])

def plot(lims=(0., 0.8)):
  
  fig, axes = plt.subplots(2, 4, False, False, figsize=(24,8))
  
  for i, row in enumerate(axes):
    for j, ax in enumerate(row):
      reward = rewards[i]
      N = Ns[j]
      for coda in agents:
        plot_result(reward, N, coda, ax = ax)
      if i==1 and j == 3:
        ax.legend(loc='lower right', fontsize='x-large')
      ax.set_xlabel('Thousands of steps')
      ax.set_ylabel('Average Reward')
      ax.set_title(f"{reward} reward, place {N} task")
      ax.set_xlim(0, M)
      ax.set_ylim(*lims)
  
  plt.tight_layout()
  plt.savefig('results_plot.png', dpi=300)
  plt.show()
  
```

```python
plot()
```

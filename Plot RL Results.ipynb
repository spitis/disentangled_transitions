```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
```

```python
td3 = np.load('results/TD3_0.npy')
ddpg= np.load('results/DDPG_0.npy')
```

```python
plt.plot(td3, label='td3')
plt.plot(ddpg, label='ddpg')
plt.legend()
plt.xlabel('2000s of steps')
plt.ylabel('Average Reward')
plt.title('Minimize Pairwise Distance Task')
```
